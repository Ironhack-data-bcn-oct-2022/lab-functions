## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python

def pangram(string):
    string= string.replace(" ", '').replace(',', '').replace('.', '')
    letters= []
    #is len(letters) == 26?
    #fill letters with unique values 
    for l in string: 
        if l not in letters:
            letters.append(l) 
    if len(letters)== 26: 
        answer= f"{string} is a pangram"
    else: answer= f"{string} is not a pangram" 
    return answer 
pangram('Waltz nymph . orquickjigsvexBud')   
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(string):
    count_lower = 0
    count_upper = 0
    count_special = 0
    count_number = 0
    lower = "abcdefghijklmnopqrstuvwxyz"
    upper = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    special = "#@!$%&()^*[]{}"
    number = '1234567890'
    for i in string:
        if i in lower:
            count_lower += 1
        elif i in upper:
            count_upper += 1
        elif i in special:
            count_special += 1
        elif i in number:
            count_number += 1
        else:
            return False
    return count_lower >= 1 and count_upper >= 1 and count_special >= 1 and count_number >= 1 and len(string) >= 8
```

```python
# This will test your function 
test_pass(check_pass)
```
